Aspect ratio refers to the proportional relationship between the width and height of an image or display screen, typically expressed as two numbers separated by a colon (e.g., 16:9). It defines the shape of the element, whether it's a photo, video, or screen. Common examples include 4:3 (older standard definition), 16:9 (widescreen), and 1:1 (square).

##### Bar plot

```python
studs = [67,89,123,190,104,65,81]
subs = ['C','C++','Java','Python','PHP','JavaScript','R']
```

```python
plt.figure(figsize=(16,9))
plt.title('Student Enrollment')
plt.xlabel('Subject Name')
plt.ylabel('Total Students')
plt.bar(subs, studs, color = 'r')
```

```python
plt.figure(figsize=(16,9))
plt.title('Student Enrollment')
plt.ylabel('Subject Name')
plt.xlabel('Total Students')
plt.barh(subs, studs, color = 'r')
```

```python
openings = [18,56,100,278,167,45,13]
```

```python
plt.figure(figsize=(16,9))
plt.title('Student Enrollment')
plt.xlabel('Subject Name')
plt.ylabel('Total Students')
plt.bar(subs, studs, color = 'r', width=0.4, align='edge', label = 'students')
plt.bar(subs, openings, color = 'cyan', width=0.4, align='center', label = 'openings')
plt.legend()
```

##### Pie chart

```python
plt.figure(figsize=(5,5))
plt.pie(studs, labels=subs, autopct='%2.2f%%');
```

```python
plt.figure(figsize=(5,5))
plt.pie(openings, labels=subs, autopct='%2.2f%%');
```
